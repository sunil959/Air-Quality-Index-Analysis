## Analysis of Air Quality of Chandigarh for the year 2016-2020

## Importing the CSV file and some libraries

```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
csv_path ="./DataSet/Chandigarh/Chandigarh.csv"

df = pd.read_csv(csv_path, parse_dates= ["Date"])

df.set_index ("Date", inplace = True)

df.head(5)
```

```python
#Initializing the parameters
pm10 = df[df["Specie"] == "pm10"]
pm25 = df[df["Specie"] == "pm25"]
so2 = df[df["Specie"] == "so2"]
no2 = df[df["Specie"] == "no2"]
o3 = df[df["Specie"] == "o3"]
co = df[df["Specie"] == "co"]
```

## Data representation for PM10:
This Data represents pm10 for all years i.e. 2016-2020

```python
import matplotlib.patches as mpatches

def bars_pm10(bars):
    if bars <= 50.0:
        return "good"
    elif bars <= 100.0 :
        return "satisfactory"
    elif bars <=250.0:
        return "moderatley polluted"
    elif bars <= 350.0:
        return "poor"
    elif bars <= 430.0:
        return "very poor" 
    else :
        return "severe"
    
count_pm10 = pm10["Value"].apply(bars_pm10).value_counts().plot(kind = "barh",color=['limegreen','sandybrown', 'forestgreen', 'lightcoral',  'indianred', 'maroon'],edgecolor='black',figsize=(10,5))
red_patch1 = mpatches.Patch(color = 'forestgreen', label='0-50 : Good')
red_patch2 = mpatches.Patch(color = 'limegreen',label='51-100 : Satisfactory')
red_patch3 = mpatches.Patch(color = 'sandybrown', label='101-250 : Moderately Polluted')
red_patch4 = mpatches.Patch(color = 'lightcoral',label='251-350 : Poor')
red_patch5 = mpatches.Patch(color = 'indianred',label='351-430: Very Poor')
red_patch6 = mpatches.Patch(color = 'maroon',label='430+: Severe')

plt.legend(handles=[red_patch1,red_patch2,red_patch3,red_patch4,red_patch5,red_patch6],bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.title("PM10 for the years 2016-2020 with parameters")
```
